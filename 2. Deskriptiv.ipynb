```python
import pandas as pd

# Load the Excel file into a DataFrame
deskriptiv = pd.read_excel('Data/Data_BA_modified.xlsx')

deskriptiv
```

```python
# Compute descriptive statistics for the specific columns
desc_stats = deskriptiv[["sag_varighed", "PrisIAlt", "Indsatslængde", "alder_ved_hændelse", "alder"]].describe()

# Select only 'count', 'mean', and 'std' rows from the descriptive statistics
filtered_stats = desc_stats.loc[['count', 'mean', 'std']]

# Round 'mean' and 'std' to 2 decimal places
filtered_stats.loc['mean'] = filtered_stats.loc['mean'].round(2)
filtered_stats.loc['std'] = filtered_stats.loc['std'].round(2)

# Transpose DataFrame so that 'count', 'mean', and 'std' become columns
transposed_stats = filtered_stats.transpose()

# Convert the filtered and formatted descriptive statistics to LaTeX format
latex_code = transposed_stats.to_latex()

print(latex_code)
```

```python
# Define the relevant columns
relevant_columns = [
    'Indsats_Familiebehandling',
    'Indsats_Kontaktperson',
    'Indsats_Pasning',
    'Indsats_Psykologisk behandling',
    'Indsats_Socialpædagogisk behandling'
]

# Generate the counts from your DataFrame called 'deskriptiv'
indsats_counts = {col: deskriptiv[col].value_counts().to_dict() for col in relevant_columns}

# Prepare the LaTeX table content
latex_table_content = "\\begin{tabular}{l|c}\n"
latex_table_content += "\\textbf{Indsats Type} & \\textbf{Antal Observationer} \\\\\\hline\n"

total_count = 0  # Initialize the total count

for col in relevant_columns:
    counts = indsats_counts[col]
    if 1 in counts:  # Only include rows with the value (1)
        latex_table_content += "{} & {} \\\\\n".format(col.replace('_', '\\_'), counts[1])
        total_count += counts[1]  # Add to the total count

# Add the total count row
latex_table_content += "\\textbf{I Alt} & \\textbf{" + str(total_count) + "} \\\\\\hline\n"

latex_table_content += "\\end{tabular}"

print(latex_table_content)
```
